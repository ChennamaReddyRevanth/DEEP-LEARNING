Activation functions play a crucial role in artificial neural networks by introducing nonlinearity into the model. They determine the output of a node or neuron, allowing the network
to learn complex patterns and relationships. Here are some commonly used activation
functions along with their implementations in Python using the NumPy library:

```python
#1.Sigmoid Activation Function
# The sigmoid function squashes its input to a range between 0 and 1
import numpy as np
def sigmoid(x):
  return 1/(1+np.exp(-x))

#Example usage:
x_values=np.linspace(-5,5,100)
y_values=sigmoid(x_values)

#plot the sigmoid functoin
from matplotlib import pyplot as plt
plt.plot(x_values,y_values,label='sigmoid')
plt.title("Sigoid Activation Function")
plt.xlabel("Input")
plt.ylabel("output")
plt.legend()
plt.show()
```

```python
#2. Hyperbolic Tangent (tanh) Activation Function:
# The tanh function squashes its input to a range between -1 and 1

def tanh(x):
  return np.tanh(x)
y_values_tanh=tanh(x_values)
plt.plot(x_values,y_values_tanh,label='Hyperbolic Tangent')
plt.title("tanh Activation Function")
plt.xlabel("Input")
plt.ylabel("output")
plt.legend()
plt.show()
```

```python
#3.Rectified Linear Unit (ReLU) Activation Function:
#The ReLU function outputs the input for positive values and zero for negative values
def relu(x):
  return np.maximum(0,x)

y_values_relu=relu(x_values)

#plot the ReLU function
plt.plot(x_values,y_values_relu,label="ReLU",color='green')
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.show()
```

```python
#4.Leaky Rectified Linear UNit (Leaky ReLU) Activation Function:
#Leaky ReLU allows a small,non-zerogradient when the input is negative

def leaky_relu(x,alpha=0.01):
  return np.where(x>0,x,alpha*x)

y_values_leaky=leaky_relu(x_values)

#plot the leaky ReLU function
plt.plot(x_values,y_values_leaky,label="Leaky ReLU")
plt.title("Leaky Rectified Linear Unit")
plt.xlabel("Input")
plt.ylabel("output")
plt.legend()
plt.show()
```

```python
def softmax(x):
  exp_x=np.exp(x-np.max(x,axis=-1,keepdims=True))
  return exp_x/np.sum(exp_x,axis=-1,keepdims=True)

scores=np.array([2.0,1.0,0.1])
probs=softmax(scores)
print("Softmax Probabilities :",probs)
```
